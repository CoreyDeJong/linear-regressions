```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python

# salary = pd.read_csv("./salary.csv")
beers = pd.read_csv("./beers.csv").dropna()
# salary = pd.read_csv("./salary_plus_snacks.csv")
print(f"Data Shape: {beers.shape}")
print(beers.head())
```

```python
print(beers.columns)
```

```python
#Selecting Corresponding Features
X = beers['abv'].values
# X = salary['Snacks'].values
y = beers['ibu'].values
X
```

```python
X=X.reshape(-1,1)
X
```

```python
# X_feature - Years of Experience
# y_output - Salary in $
x_train, x_test, y_train, y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=100)
print(f"X_train shape {x_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shap {x_test.shape}")
print(f"y_test shape {y_test.shape}")
print(y_test)
```

```python
%matplotlib inline
plt.scatter(x_train,y_train,color='red')
plt.xlabel('ABV')
plt.ylabel('IBU')
plt.title('ABV vs IBU')
plt.show()
```

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
y_predict = lm.predict(x_test)
print(f"Train accuracy {round(lm.score(x_train,y_train)*100,2)} %")
print(f"Test accuracy {round(lm.score(x_test,y_test)*100,2)} %")
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_test,y_predict)
plt.xlabel("ABV")
plt.ylabel("IBU")
plt.title("Alcohol coorelation to Bitterness")
plt.plot
```
